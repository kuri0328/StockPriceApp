```python
days = 20
tickers = {
    'apple':'AAPL',
    'microsoft':'MSFT',
    'google':'GOOGL',
    'netflix':'NFLX',
    'amazon':'AMZN',
    'meta':'meta'
}

df = pd.DataFrame()
for company in tickers.keys():
#company = 'apple'
    tkr = yf.Ticker(tickers[company])
    hist = tkr.history(period=f'{days}d')
    hist.index = hist.index.strftime('%d %B %Y')
    hist = hist[['Close']]
    hist.columns = [company]
    hist = hist.T
    hist.index.name = 'Name'
    df = pd.concat([df, hist])
```

```python
def get_data(days, tickers):
    df = pd.DataFrame()
    for company in tickers.keys():
    #company = 'apple'
        tkr = yf.Ticker(tickers[company])
        hist = tkr.history(period=f'{days}d')
        hist.index = hist.index.strftime('%d %B %Y')
        hist = hist[['Close']]
        hist.columns = [company]
        hist = hist.T
        hist.index.name = 'Name'
        df = pd.concat([df, hist])
    return df
```

```python
days = 20
tickers = {
    'apple':'AAPL',
    'microsoft':'MSFT',
    'google':'GOOGL',
    'netflix':'NFLX',
    'amazon':'AMZN',
    'meta':'meta'
}
get_data(days, tickers)
```

```python
companies = ['google', 'microsoft']
data = df.loc[companies]
data
```

```python
data.sort_index()
```

```python
data = data.T.reset_index()
data.head()
```

```python
data = pd.melt(data, id_vars=['Date']).rename(columns={'value':'Stock Prices(USD)'}) #生データにばらす
data
```

```python
ymin, ymax = 300, 350
chart = (
    alt.Chart(data)
    .mark_line(opacity=0.8,clip=True)
    .encode(
        x="Date:T",
        y=alt.Y("Stock Prices(USD):Q", stack=None, scale=alt.Scale(domain=[ymin,ymax])),
        color='Name:N'
    )
)
```
